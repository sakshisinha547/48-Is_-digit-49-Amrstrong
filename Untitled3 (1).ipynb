```python
def is_digit(text:str) -> int:
    for char in text:
        if char <"0" or char >"9":
            return False
        return True

if __name__ == "__main__":
    try:
        text  = "7344794"
        print(is_digit(text))
    except Exception as e:
        print(e)
```

```python
def Armstrong(number :int):
    temp = number
    digit_count = 0 
    while temp > 0:
        digit_count = digit_count +1
        temp = temp // 10

    total  = 0
    temp = number
    while temp > 0:
        digit = temp % 10
        total = total +digit**digit_count
        temp = temp //10
    return total == number

if __name__ == "__main__":
    try:
        number = 153
        print(Armstrong(number))
    except Exception as e:
        print(e)
    
        
```
